#### Imports

```python
from xml.etree.ElementTree import ElementTree as et
from xml.etree.ElementTree import fromstring
import urllib.request
import bs4 as bs
```

#### Reuters XML Feed

```python
source = urllib.request.urlopen('http://feeds.reuters.com/reuters/topNews').read()
```

```python
tree = fromstring(source.decode('utf-8'))
```

```python
root = tree.getchildren()
firstchild = root[0]
```

```python
itemlist = firstchild.findall('item')
len(itemlist)
for items in itemlist:
    print("Title: ",items[0].text,"\n")
    #source = urllib.request.urlopen(items[2].text).read()
    #soup = bs.BeautifulSoup(source,'lxml')
    #story = soup.find_all("p", {"class": "StandardArticleBody_body"})
    print(itemlist.index(items)+1,items[2].text,"\n")
    #print("Body text : ",source,"\n")
```

#### Daily Mail XML Feed

```python
daily_mail_source = urllib.request.urlopen('https://www.dailymail.co.uk/home/index.rss').read()
tree = fromstring(daily_mail_source.decode('utf-8'))
root = tree.getchildren()
firstchild = root[0]
```

```python
itemlist = firstchild.findall('item')
len(itemlist)
for items in itemlist:
    print("Title: ",items[0].text,"\n")
    #source = urllib.request.urlopen(items[2].text).read()
    #soup = bs.BeautifulSoup(source,'lxml')
    #story = soup.find_all("p", {"class": "StandardArticleBody_body"})
    print(itemlist.index(items)+1,items[2].text,"\n")
    #print("Body text : ",source,"\n")
    if itemlist.index(items) > 18:
        break
```

#### Sun XML Feed

```python
sun_source = urllib.request.urlopen('https://www.thesun.co.uk/feed/').read()
sun_soup = bs.BeautifulSoup(sun_source,'lxml')
sun_titles = sun_soup.find_all("span", {"class": "rail__item-sub"})
```

```python
sun_titles_list = []
for title in sun_titles:
    sun_titles_list.append(title.getText().strip())

sun_titles_list = list(set(sun_titles_list))
for i in sun_titles_list:
    print(sun_titles_list.index(i), " \n",i)
```
